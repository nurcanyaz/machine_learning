```python
import matplotlib.pyplot as plt

movie_year = veri[veri['type'] == 'Movie']['listed_in'].str.split(', ').explode()

son_on_yıl = movie_year.value_counts()


plt.figure(figsize=(10, 6))
son_on_yıl.plot(kind='bar')
plt.title('Movie Kategorilerine Göre Film Sayısı')
plt.xlabel('Kategori')
plt.ylabel('Film Sayısı')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

movie_kategoriler = veri[veri['type'] == 'Movie']['listed_in'].str.split(', ').explode()

en_populer_10_kategori = movie_kategoriler.value_counts().nlargest(10)


plt.figure(figsize=(10, 6))
en_populer_10_kategori.plot(kind='bar')
plt.title('En populer 10 film kategorisi')
plt.xlabel('Kategori')
plt.ylabel('Film Sayısı')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
